```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mplt
%matplotlib inline
```

```python
df = pd.read_csv('2018_maternal_mortality_us.csv')
print(df)
```

```python
plt.plot(df['Value'], color='brown')
plt.ylim([-0.1, 175])
plt.ylabel('Number of Deaths')
plt.title('Maternal Mortality Rates')
plt.show()
```

```python
plt.scatter(x=df['Score'], y=df['Value'], color='red')
plt.ylabel('Number of Deaths')
plt.title('Maternal Mortality Rates')
plt.show()
```

```python
# How do I create a histogram, if I am not looking for a randomized result?
value = np.random.normal(10, 5, 1000)

# Plot them as a histogram.
plt.hist(value)
plt.title('Maternal Mortality')
plt.show()
```
